```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titles = ["Sequence","Mcg","Gvh","Lip","Chg","Aac","Alm1","Alm2","Class"]
data = pd.read_table("http://archive.ics.uci.edu/ml/machine-learning-databases/ecoli/ecoli.data", 
                     header= None, 
                     sep="\s+", 
                     names= titles)
```

### 2. Summarize each field in the data, i.e. mean, average etc.
### 4. Group data by their classes.
    (a) Give average mcg, average gvh and average alm2 of each group.
    (b) Plot for each field (mcg, gvh, alm2). Use a boxplot that visualizes the statistical information
    about them.
    (c) For each class, draw a prediction line for mcg vs chg



```python
Class_Types = data.Class.unique()
print("The total number of clases are: ",len(Class_Types))
print("Bellow are the statistics of each class")

n = 0
while n < len(Class_Types):
    print ('Data of Class "',Class_Types[n], '" has this statistics:')
    f= data[data.Class == Class_Types[n]] #Filtering the information by the starting Class_Type[0]= "cp" 
    del f['Lip']
    del f['Aac']
    del f['Alm1'] #Removed columns, to have a cleaner view of the .describe() function
    print (f.describe()) #Show the statistical data of the remaining DataFrame
    plt.boxplot([f['Mcg'],f['Gvh'],f['Alm2']], labels = ["Mcg","Gvh","Alm1"])
    plt.title(["Boxplot of Bacteria Class Type", Class_Types[n]]) #Calling the name of the Class_Type[0]
    plt.show()
    print ("")
    x1 = np.array(f['Mcg'])
    y = np.array(f['Chg'])
    A = np.vstack([x1, np.ones(len(x1))]).T
    b1, b0 = np.linalg.lstsq(A, y)[0]
    print ("Linear Regresion Equation is Chg = Mcg*",b1,"+ ",b0 )
    plt.plot(x1, y, 'bo', label='Original data', markersize=4)
    plt.plot(x1, b1*x1 + b0, 'r', label='Fitted line')
    plt.legend()
    plt.xlabel('Mcg')
    plt.ylabel('Chg')
    plt.title('Regression Line of Mcg vs Chg')
    plt.show()
    print ("")
    n = n + 1
  
        
```
